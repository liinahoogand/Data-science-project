# Introduction to the dataset
Dataset we are using in this project consists of  4424 rows and 35 attributes. This dataset is created from a higher education institution (acquired from several disjoint databases) related to students enrolled in different undergraduate degrees, such as agronomy, design, education, nursing, journalism, management, social service, and technologies. Dataset includes information known at the time of student enrollment (academic path, demographics, and social-economic factors) and the students' academic performance at the end of the first and second semesters. The data refer to records of students enrolled between the academic years 2008/2009 to 2018/2019.

# Some basic data analysis

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data['Target'] = data['Target'].map({
    'Dropout':0,
    'Enrolled':1,
    'Graduate':2
})
```

```python
data.head(5)
```

```python
gender_counts = data['Gender'].map({0: 'Female', 1: 'Male'}).value_counts()
plt.pie(gender_counts, labels=gender_counts.index,
        autopct=lambda p: '{:.0f}\n({:.1f}%)'.format(p * sum(gender_counts) / 100, p),
        startangle=90, colors=plt.cm.Pastel1.colors)
plt.title('Distribution of Gender')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(data['Age at enrollment'], bins=50, edgecolor='black', color='lightblue')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Ages of enrollment')
plt.show()
```

```python
target_labels = {0: 'Dropout', 1: 'Enrolled', 2: 'Graduate'}

plot_data = data.copy()
plot_data['Target'] = plot_data['Target'].map(target_labels)

filtered_data = plot_data[plot_data['Age at enrollment'] <= 40]

sns.set_palette("Pastel1")
sns.countplot(x='Age at enrollment', hue='Target', data=filtered_data, hue_order=['Dropout', 'Enrolled', 'Graduate'])

plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Count Plot of Target by Age at Enrollment (up to 40 years old)')

plt.legend(title='Target', title_fontsize='13')
plt.show()
```

```python
filtered_data = data[data['Age at enrollment'] <= 40]

target_counts_by_age = filtered_data.groupby(['Age at enrollment', 'Target']).size().unstack(fill_value=0)

target_labels = {0: 'Dropout', 1: 'Enrolled', 2: 'Graduate'}
target_counts_by_age.columns = target_counts_by_age.columns.map(target_labels)

percentage_values_by_age = target_counts_by_age.div(target_counts_by_age.sum(axis=1), axis=0) * 100

percentage_values_by_age.plot(kind='bar', colormap='Pastel1', stacked=True)
plt.xlabel('Age at Enrollment')
plt.ylabel('Percentage')
plt.title('Percentage Distribution of Target for Ages up to 40')
plt.legend(title='Target', title_fontsize='13')
plt.show()
```

```python
student_target_counts = data['Target'].map({0: 'Dropout', 1: 'Enrolled', 2: 'Graduate'}).value_counts()
colors = plt.cm.Pastel1.colors

plt.pie(student_target_counts, labels=student_target_counts.index,
        autopct=lambda p: '{:.0f}\n({:.1f}%)'.format(p * sum(student_target_counts) / 100, p),
        startangle=90, colors=colors)

plt.title('Distribution of Target Variable')
plt.show()
```

```python
gender_labels = {0: 'Female', 1: 'Male'}
target_labels = {0: 'Dropout', 1: 'Enrolled', 2: 'Graduate'}

plot_data = data.copy()
plot_data['Gender'] = plot_data['Gender'].map(gender_labels)
plot_data['Target'] = plot_data['Target'].map(target_labels)

count_data = plot_data.groupby(['Gender', 'Target']).size().reset_index(name='Count')

sns.barplot(x='Gender', y='Count', hue='Target', data=count_data, palette='Pastel1')

plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Distribution of Target by Gender')

plt.show()
```

```python
gender_labels = {0: 'Female', 1: 'Male'}
target_labels = {0: 'Dropout', 1: 'Enrolled', 2: 'Graduate'}

plot_data = data.copy()
plot_data['Gender'] = plot_data['Gender'].map(gender_labels)
plot_data['Target'] = plot_data['Target'].map(target_labels)

total_count = len(plot_data)

count_data = plot_data.groupby(['Gender', 'Target']).size().reset_index(name='Count')

count_data['Percentage'] = count_data.groupby('Gender')['Count'].transform(lambda x: x / x.sum() * 100)

sns.barplot(x='Gender', y='Percentage', hue='Target', data=count_data, palette='Pastel1')

plt.xlabel('Gender')
plt.ylabel('Percentage')
plt.title('Percentage of Target by Gender')

plt.show()
```

```python
percentage_values = (
    data.groupby(['Daytime/evening attendance', 'Target']).size() /
    data.groupby('Daytime/evening attendance').size()
) * 100

percentage_values = percentage_values.unstack(fill_value=0)

sns.set_palette("Pastel1")
ax = percentage_values.plot(kind='bar', stacked=True)

plt.xlabel('Daytime/evening attendance')
plt.ylabel('Percentage')
plt.title('Percentage Distribution of Target by Daytime/Evening Attendance')

plt.xticks(range(len(percentage_values.index)), ['Night', 'Day'], rotation=0)

plt.legend(title='Target', title_fontsize='13', labels=['Dropout', 'Enrolled', 'Graduate'], loc='upper right')
plt.show()
```

# Creating predicting model

To assess the potential impact of columns with low correlation on the predictive performance of the model, we will compare the model accuracies between the **original dataset** and a modified dataset. In the **modified dataset**, we will exclude a columns that may not contribute significant predictive power to the target variable 'Target'.
Additionally, the predictive performance of **various models** will be evaluated and compared. The best hyperparameters for best performing model will be determined through a **randomized search** process, as using grid search turned out to be too costly.
To detect and reduce overfitting, also to evaluate performance of the model, **cross-validation** will be used.

Creating modified dataset.

```python
modified_data = data.copy()
modified_data = modified_data.drop(columns=['Father\'s qualification', 'Nacionality', 'International',
'Educational special needs', 'Course', 'Unemployment rate', 'Inflation rate', 'Mother\'s qualification',
'GDP'], axis=1)
modified_data.head()
```

Dividing data into training and testing sets.

```python
from sklearn.model_selection import train_test_split
#Modified dataset
X = modified_data.drop('Target', axis=1)
y = modified_data['Target']
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)#Size of the Test set is 20% of the whole dataset(4424 rows) as this allows model to be trained on a big amount of data while still having a reasonable amount for testing.
```

```python
#Original dataset
X = data.drop('Target', axis=1)
y = data['Target']
X_trainII,X_testII,y_trainII,y_testII = train_test_split(X,y,test_size=0.2,random_state=0)
```

Chosen models for comparison: **Random Forest Classifier**, **Logistic Regression** and  **Support Vector Machine**.

*Random Forest Classifier*

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

rf = RandomForestClassifier(random_state=42)#commonly used value, which worked better than 2.
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)
acc = accuracy_score(y_test,y_pred)
print('Accuracy of RF model on modified dataset: ', acc)

rf.fit(X_trainII, y_trainII)
y_pred = rf.predict(X_testII)
acc = accuracy_score(y_testII,y_pred)
print('Accuracy of RF model on original dataset: ', acc)
```

*Logistic Regression*

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(random_state=42, solver='liblinear', max_iter=1000)#using the solver parameter and adjusting the max_iter parameter resolves the ConvergenceWarning.
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
acc = accuracy_score(y_test,y_pred)
print('Accuracy of LR model on modified dataset: ', acc)

lr.fit(X_trainII, y_trainII)
y_pred = lr.predict(X_testII)
acc = accuracy_score(y_testII,y_pred)
print('Accuracy of LR model on original dataset: ', acc)
```

*Support Vector Machine(Support Vector Classification) - Linear and RBF kernel*

```python
from sklearn.svm import SVC

svc = SVC(kernel='linear',probability=True)
svc.fit(X_train, y_train)
y_pred = svc.predict(X_test)
acc = accuracy_score(y_test,y_pred)
print('Accuracy of SVC(linear) model on modified dataset: ', acc)

svc.fit(X_trainII, y_trainII)
y_pred = svc.predict(X_testII)
acc = accuracy_score(y_testII,y_pred)
print('Accuracy of SVC(linear) model on original dataset: ', acc)

svc = SVC(kernel='rbf',probability=True)
svc.fit(X_train, y_train)
y_pred = svc.predict(X_test)
acc = accuracy_score(y_test,y_pred)
print('Accuracy of SVC(RBF) model on modified dataset: ', acc)

svc.fit(X_trainII, y_trainII)
y_pred = svc.predict(X_testII)
acc = accuracy_score(y_testII,y_pred)
print('Accuracy of SVC(RBF) model on original dataset: ', acc)
```

Best results were delivered by Random Forest classifier with accuracy of **0.7954802259887006** on modified dataset and **0.8033898305084746** on original dataset.

# Hyphotesis

Although correlation analysis showed that correlation between Father's and Mother's qualification with Target value is almost zero, we decided to test hyphotesis, that these factors to affect academic succes of a student. For testing, we will perform Chi-squared Test, as it is the most suitable for examining the relationship between two categorical variables.

**Null Hypothesis (H0)**:

There is no significant association between the qualification of parents and the academic success of their children.

**Alternative Hypothesis (H1)**:

There is a significant association between the qualification of parents and the academic success of their children.

```python
from scipy.stats import chi2_contingency
from sklearn.metrics import confusion_matrix, classification_report
import numpy as np
```

For optimal visualization of the data, we grouped 34 qualifications into 5 main categories: **Elementary School, High School, Professional course, High education, No education**.

```python
def categorize_education_level(education_level):
    if education_level in [7,8,20,21,18,9,27,28,14,26,12,10]:
        return '1'
    elif education_level in [1,11,15,17,19]:
        return '2'
    elif education_level in [6,13,16,22,23,29,31,32]:
        return '3'
    elif education_level in [3,5,2,30,33,34]:
        return '4'
    else:
       return '0'

data_qualification = data.copy()
data_qualification["Mother's qualification"] = data_qualification["Mother's qualification"].apply(categorize_education_level)
data_qualification["Father's qualification"] = data_qualification["Father's qualification"].apply(categorize_education_level)

```

```python
qualification_names = {
    '1': 'Elementary School',
    '2': 'High School',
    '3': 'Professional course',
    '4': 'High education',
    '0': 'No education',
}
data_qualification["Mother's qualification"] = data_qualification["Mother's qualification"].map(qualification_names)
data_qualification["Father's qualification"] = data_qualification["Father's qualification"].map(qualification_names)
```

Distribution of Target values by categorized qualification of the parents.

```python

def plot_qualification_distribution(data, qualification_type):
    total_counts = data.groupby(f"{qualification_type}'s qualification")["Target"].count()
    percentage_data = data.groupby([f"{qualification_type}'s qualification", "Target"]).size() / total_counts * 100
    percentage_data = percentage_data.reset_index(name='Percentage')

    target_labels = {0: 'Dropout', 1: 'Enrolled', 2: 'Graduate'}
    percentage_data['Target'] = percentage_data['Target'].map(target_labels)

    sns.barplot(x=f"{qualification_type}'s qualification", y='Percentage', hue='Target', data=percentage_data, palette='Pastel1')

    plt.xlabel(f"{qualification_type}'s qualification")
    plt.ylabel('Percentage')
    plt.title(f'Distribution of Dropout and Graduate Cases by {qualification_type}\'s Qualification')
    plt.xticks(rotation=45, ha='right', fontsize=10)
    plt.legend(title='Target', title_fontsize='13', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.show()

plot_qualification_distribution(data_qualification, "Mother")
plot_qualification_distribution(data_qualification, "Father")
```

Testing the hyphotesis with **Chi-squared Test**.

```python
contingency_table = pd.crosstab(data["Father's qualification"], data["Target"])
chi2 = chi2_contingency(contingency_table)
print('Mother\'s qualification.')
print(f'Chi-squared value: {chi2.statistic}\nP-value: {chi2.pvalue}')
```

```python
contingency_table = pd.crosstab(data["Mother's qualification"], data["Target"])
chi2 = chi2_contingency(contingency_table)
print('Father\'s qualification.')
print(f'Chi-squared value: {chi2.statistic}\nP-value: {chi2.pvalue}')
```

**Meaning of the values:**

The high chi-squared value indicates a significant association between
"Mother's qualification" and the "Target" variable. The very low p-value (3.175e-19) implies **strong evidence against the null hypothesis** supporting the conclusion that "Mother's qualification" is significantly associated with the academic outcome represented by the "Target" variable.

Similarly, for "Father's qualification," the chi-squared value of 217.9521 and the extremely low p-value (5.8238e-21) indicate a **statistically significant association** with the "Target" variable.

Here is a visualization of expected and observed results for each target found during Chi-squared Test.

```python
def plot_observed_and_expected(data, qualification_type,c1,c2):
    contingency_table = pd.crosstab(data[f"{qualification_type}"], data["Target"])

    observed_row_totals = contingency_table.sum(axis=1)
    observed_col_totals = contingency_table.sum(axis=0)
    total_observed = observed_row_totals.sum()

    expected_frequencies = np.outer(observed_row_totals, observed_col_totals) / total_observed
    expected_frequencies = pd.DataFrame(expected_frequencies, columns=contingency_table.columns, index=contingency_table.index)

    width = 0.35

    fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(15, 6), sharey=True)

    for i, target_category in enumerate(contingency_table.columns):
        ax = axes[i]
        observed_bars = ax.bar(np.arange(len(contingency_table.index)), contingency_table[target_category], width, label='Observed', color=sns.color_palette('Pastel1')[c1])
        expected_bars = ax.bar(np.arange(len(expected_frequencies.index)) + width, expected_frequencies[target_category], width, label='Expected', alpha=0.5, color=sns.color_palette('Pastel1')[c2])

        ax.set_xticks(np.arange(len(contingency_table.index)) + width / 2)
        ax.set_xticklabels(contingency_table.index, rotation=45, ha='right')
        ax.set_xlabel(f"{qualification_type}'s qualification")
        ax.set_ylabel('Frequency')
        ax.set_title(f'Observed and Expected Frequencies for Target={target_category}')
        ax.legend()

    plt.tight_layout()
    plt.show()
plot_observed_and_expected(data_qualification, "Father's qualification",1,2)
plot_observed_and_expected(data_qualification, "Mother's qualification",0,4)
```

In summary, we accept the **alternative hypothesis**, affirming a meaningful association between parental qualifications and a child's academic success, even though this relationship may not be indicated by correlation between these factors.
